```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
from matplotlib import transforms
import ipywidgets as widgets
from IPython.display import HTML

plt.style.use('noelle.mplstyle')

# inline or notebook or qt
%matplotlib qt 

# create a figure and axes
fig, axs = plt.subplots(figsize=(8,8),nrows=2,ncols=2)
axs[0][0].set_xlim((-1,1))
axs[0][0].set_ylim((-1,1))            
axs[0][0].set_xlabel('Realteil')
axs[0][0].set_ylabel('Imagin√§rteil')
circle = 0.5*np.exp(1j*np.arange(101)*2*np.pi/100)
axs[0][0].plot(circle.real,circle.imag,'k--')
axs[0][0].set_aspect('equal','box')

axs[1][0].set_xlim((-1,1))
axs[1][0].set_ylim((-2,0))                    
axs[1][0].set_xlabel('Amplitude')
axs[1][0].set_ylabel('Zeit')
axs[1][0].set_aspect('equal','box')

axs[0][1].set_xlim((0,2))
axs[0][1].set_ylim((-1,1))                    
axs[0][1].set_xlabel('Zeit')
axs[0][1].set_ylabel('Amplitude')
axs[0][1].set_aspect('equal','box')

# create static objects
txt_title = axs[0][0].set_title('')
# exp cw
arr1 = axs[0][0].arrow([], [], [], [], color='b', width=0.02,length_includes_head=True, label='$0.5exp(j\omega t)$')
# exp ccw
arr2 = axs[0][0].arrow([], [], [], [], color='g', width=0.02,length_includes_head=True, label='$0.5exp(-j\omega t)$')
arr3 = axs[0][0].arrow([], [], [], [], color='g', width=0.02, alpha=0.3,length_includes_head=True)
# cos
arr4 = axs[0][0].arrow([], [], [], [], color='orange', width=0.02,length_includes_head=True)
# -exp ccw
arr5 = axs[0][0].arrow([], [], [], [], color='g', width=0.02, alpha=0.3,length_includes_head=True)
# sin
arr6 = axs[0][0].arrow([], [], [], [], color='red', width=0.02,length_includes_head=True)
# verical line
l1, = axs[0][0].plot([],[],color='orange',ls='--')
# horizontal line
l2, = axs[0][0].plot([],[],color='red',ls='--')
axs[0][0].legend(fontsize=10, loc='upper right')

# rotate 2nd plot 
# https://copyprogramming.com/howto/how-can-i-rotate-a-matplotlib-plot-through-90-degrees?utm_content=cmp-true
base = axs[1][0].transData
rot = transforms.Affine2D().rotate_deg(-90)

cos1, = axs[1][0].plot([],[],color='orange',ls='-',label='$cos(\omega t)$', transform=rot+base)
sin1, = axs[0][1].plot([],[],color='red',ls='-',label='$sin(\omega t)$')
x = np.arange(0,2,0.002)
ycos = np.cos(2*np.pi*x)
ysin = np.sin(2*np.pi*x)
axs[1][0].legend(loc='upper right',fontsize=10)

# animation function. This is called sequentially
def drawframe(n):
    e1 = 0.5*np.exp(1j*2*np.pi * n / 50)
    arr1.set_data(x=0,y=0,dx=e1.real,dy=e1.imag)
    arr2.set_data(x=0,y=0,dx=e1.real,dy=-e1.imag) 
    arr3.set_data(x=e1.real,y=e1.imag,dx=e1.real,dy=-e1.imag)  
    arr4.set_data(x=0,y=0,dx=2*e1.real,dy=0)  
    l1.set_data([2*e1.real,2*e1.real],[-1,0])
    arr5.set_data(x=e1.real,y=e1.imag,dx=-e1.real,dy=e1.imag)
    arr6.set_data(x=0,y=0,dx=0,dy=2*e1.imag) 
    l2.set_data([0,1],[2*e1.imag,2*e1.imag])

    cos1.set_data(x,np.roll(ycos,n*10))
    sin1.set_data(x,-np.roll(ysin,n*10))

    return (arr1,arr2,arr3,arr4,l1,cos1,arr5,arr6,l2,sin1)

from matplotlib import animation

# blit=True re-draws only the parts that have changed.
anim = animation.FuncAnimation(fig, drawframe, frames=50, interval=50, blit=True)

plt.show()
# HTML(anim.to_html5_video())

```
